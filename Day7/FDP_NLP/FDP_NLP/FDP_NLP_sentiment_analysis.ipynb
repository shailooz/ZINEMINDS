```python
import pandas as pd
import numpy as np
df= pd.read_csv('movie_reviews.csv')
df.head(10)
```

```python
summary = df.Summary.values
labels = df.Sentiment.values
```

```python
print(summary)
```

```python
print(labels)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# tokenizer with a maximum vocabulary size of 1000
tokenizer = Tokenizer(num_words=1000)

# Fit the tokenizer on the text data
tokenizer.fit_on_texts(summary)

# Convert text data to sequences of integers
encoded_docs = tokenizer.texts_to_sequences(summary)

print(encoded_docs)
```

```python
# Determine the vocabulary size
vocab_size = len(tokenizer.word_index) + 1  # Adding 1 to account for the reserved 0 index
print(vocab_size)

# Pad sequences to a fixed length of 200
padded_sequence = pad_sequences(encoded_docs, maxlen=200)
```

```python
# Build the  RNN model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Dropout, Embedding, SpatialDropout1D

embedding_vector_length = 32

model_1 = Sequential()
model_1.add(Embedding(vocab_size, embedding_vector_length, input_length=200))
model_1.add(SpatialDropout1D(0.25))  # prevent overfitting in neural networks.
model_1.add(SimpleRNN(50))
# model.add(Dropout(0.2))  # You can uncomment this line if needed
model_1.add(Dense(1, activation='sigmoid'))

model_1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model_1.summary())
```

```python
history = model_1.fit(padded_sequence,labels,validation_split=0.2, epochs=5, batch_size=32)
```

```python
test_word ="This is very bad"
tw = tokenizer.texts_to_sequences([test_word])
tw = pad_sequences(tw,maxlen=200)
predictions = model_1.predict(np.array(tw))
print(predictions)
```

```python
# Build the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense, Dropout
from tensorflow.keras.layers import SpatialDropout1D
from tensorflow.keras.layers import Embedding
embedding_vector_length = 32
model = Sequential()
model.add(Embedding(vocab_size, embedding_vector_length,
                                     input_length=200) )
# model.add(SpatialDropout1D(0.25)) # prevent overfitting in neural networks.
model.add(LSTM(50))
# model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',
                           metrics=['accuracy'])
print(model.summary())
```

```python
history = model.fit(padded_sequence,labels,validation_split=0.2, epochs=5, batch_size=32)
```

```python
test_word ="This is very bad"
tw = tokenizer.texts_to_sequences([test_word])
tw = pad_sequences(tw,maxlen=200)
predictions = model.predict(np.array(tw))
print(predictions)
```

```python
# Build the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout, Embedding, SpatialDropout1D

embedding_vector_length = 32

model_2 = Sequential()
model_2.add(Embedding(vocab_size, embedding_vector_length, input_length=200))
model_2.add(SpatialDropout1D(0.25))  # prevent overfitting in neural networks.
model_2.add(GRU(50))
# model_2.add(Dropout(0.2))  # You can uncomment this line if needed
model_2.add(Dense(1, activation='sigmoid'))

model_2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model_2.summary())

```

```python
history = model_2.fit(padded_sequence,labels,validation_split=0.2, epochs=5, batch_size=32)
```

```python
test_word ="This is very bad"
tw = tokenizer.texts_to_sequences([test_word])
tw = pad_sequences(tw,maxlen=200)
predictions = model_2.predict(np.array(tw))
print(predictions)
```
